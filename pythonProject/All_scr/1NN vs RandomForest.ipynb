```python
import sklearn.datasets
import matplotlib.pyplot as plt
import sklearn.metrics
from matplotlib.colors import ListedColormap
from sklearn import model_selection, datasets, metrics, tree, ensemble, 
import numpy as np
%pylab inline
```

```python
digits = datasets.load_digits()
```

```python
X, y = digits.data, digits.target
```

```python
dataset_size = len(X)
train_size = int(0.75*dataset_size)
train_size
```

```python
X_train = X[:train_size]
y_train = y[:train_size]
```

```python
X_test, y_test = X[train_size:dataset_size], y[train_size:dataset_size]
```

```python
def euclidian_metric(x, y):
    return np.sqrt( np.sum((x - y)**2) )
```

```python
y_pred_knn = []
for test_value in X_test:
    ind_min_metric = 0
    min_metric = euclidian_metric(test_value, X_train[0])
    
    for index, train_value in enumerate(X_train):
        metric = euclidian_metric(test_value, train_value)
        if metric < min_metric:
            min_metric = metric
            ind_min_metric = index
            print(index)
            
    y_pred_knn.append(y_train[ind_min_metric])
```

```python
knn_err_rate = 1 - sklearn.metrics.accuracy_score(y_test, y_pred_knn)
print('1nn classifier error: ' + str(knn_err_rate))
```

```python
with open('answer1.txt', 'w') as fout:
    fout.write(str(knn_err_rate))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=1000)
rf_clf.fit(X_train, y_train)
```

```python
#Random forest prediction
y_pred_rf = rf_clf.predict(X_test)

#Accuracy
rf_err_rate = 1 - sklearn.metrics.accuracy_score(y_test, y_pred_rf)
print('Random forest classifier error: ' + str(rf_err_rate))
```
