```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import expon
%matplotlib inline
```

```python
# Задаем экспоненциальное распределение
exp_rv = sts.expon()

# Получаем выборку из 1000 значений
sample = exp_rv.rvs(1000)

# Задаем область определения случайной величины и на нем теоретическую кривую плотности вероятности
x = np.linspace(0, 6, 1000)
pdf = exp_rv.pdf(x)

# Строим теоретическую кривую распределения и гистограмму выборки
fig=plt.figure(figsize=(8, 6), dpi= 80)
plt.plot(x, pdf, linewidth=2, label='theoretical PDF')
plt.hist(sample, bins=20, density = True, label='sample hist')
plt.ylabel('$PDF+HIST$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

```python
import math
# Создаем выборку из 1000 выборочных средних для объема n = 5 
sample_mean_5 = [np.mean(exp_rv.rvs(5)) for i in range(1000)]

# Строим гистограмму из полученных значений
fig=plt.figure(figsize=(8, 6), dpi= 80)
plt.hist(sample_mean_5, bins=20, density = True, label='mean sample hist')

# Задаем область определения выборочных средних
# На ней задаем нормальное распределение со средним значением 1 и дисперсией 1/n, n = 5
x = np.linspace(0, 4, 1000)
norm_rv = sts.norm(1, 1/np.sqrt(5))
pdf = norm_rv.pdf(x)

# Строим график полученного нормального распределения

plt.plot(x, pdf, "r-", linewidth=5, label='theoretical PDF', alpha=0.5)
plt.ylabel('$PDF+HIST$')
plt.xlabel('$MeanSample$')
plt.legend(loc='upper right')


std = exp_rv.std() # Среднеквадратическое отклонение
Dx = std**2
sigma = math.sqrt(Dx/5)
print(Dx, std, sigma)
```
