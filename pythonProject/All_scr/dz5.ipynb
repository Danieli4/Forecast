```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import statsmodels.stats.api as sm
import math

%matplotlib inline
%pylab inline



#ГРАФИК 1

#генерируем выборку из случайной величины, имеющей равномерное распределение на отрезке [a;b]
a=1
b=100

uniform_rv=sts.uniform(a,b-a)

ar=uniform_rv.rvs(1000)

x=np.linspace(a,b,1000)
pdf1 = uniform_rv.pdf(x)

#строим гистограмму
plt.figure(figsize(8,6))
plt.hist(ar,bins = 20, normed = True)
plt.plot(x,pdf1)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')

#ГРАФИК 2


#генерируем массив выборочных средних для выборки объемом 10
z10 = np.zeros(1000) 
for n in arange(1000):
     z10[n]=sum(uniform_rv.rvs(10))/10. #генерируем выборку оъемом 10 наблюдений и находим выброчное среднее
    
#апроксиммрующее нормальное распределение 

#среднее
EX=(b+a)/2.
#дисперсия
DX=(b-a)**2/12.

#параметры нормального распределения
mu = EX
sigma = math.sqrt(DX/10.)

# зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mu, scale=sigma)

# сгенерируем 1000 значений
norm_rv.rvs(size=1000)

# вычисляем плотность
pdf = norm_rv.pdf(x)    
    
#строим график    
    
plt.figure(figsize(8,6))
plt.hist(z10,bins = 15, normed = True)  
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')


#ГРАФИК 3

#генерируем массив выборочных средних для выборки объемом 50
z50 = np.zeros(1000) 
for n in arange(1000):
     z50[n]=sum(uniform_rv.rvs(50))/50. #генерируем выборку оъемом 10 наблюдений и находим выброчное среднее
        
  #апроксиммрующее нормальное распределение 

#среднее
EX=(b+a)/2.
#дисперсия
DX=(b-a)**2/12.

#параметры нормального распределения
mu = EX
sigma = math.sqrt(DX/50.)

# зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mu, scale=sigma)

# сгенерируем 1000 значений
norm_rv.rvs(size=1000)

# вычисляем плотность
pdf = norm_rv.pdf(x)          
#строим график    
plt.figure(figsize(8,6))
plt.hist(z50,bins = 15, normed = True)  
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')


#ГРАФИК 4

#генерируем массив выборочных средних для выборки объемом 300
z300 = np.zeros(1000) 
for n in arange(1000):
     z300[n]=sum(uniform_rv.rvs(300))/300. #генерируем выборку оъемом 10 наблюдений и находим выброчное среднее
        
  #апроксиммрующее нормальное распределение 

#среднее
EX=(b+a)/2.
#дисперсия
DX=(b-a)**2/12.

#параметры нормального распределения
mu = EX
sigma = math.sqrt(DX/300.)

# зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mu, scale=sigma)

# сгенерируем 1000 значений
norm_rv.rvs(size=1000)

# вычисляем плотность
pdf = norm_rv.pdf(x)          
#строим график    
plt.figure(figsize(8,6))
plt.hist(z300,bins = 15, normed = True)  
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')

#Вывод: при увеличении размера выборки n, распределение 
#становится все более гладким и похожим на нормальное. 
#Точность аппроксимации распределения выборочных средних нормальным с ростом n увеличивается. 
```
